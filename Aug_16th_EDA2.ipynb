```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Combine product categories into one column
df_melted = pd.melt(df, id_vars=['Age'], 
                    value_vars=['Product_Category_1', 'Product_Category_2', 'Product_Category_3'],
                    var_name='Product_Type', value_name='Product_Category')

# Drop missing values
df_melted = df_melted.dropna()

# Group and count
affinity = df_melted.groupby(['Age', 'Product_Category']).size().reset_index(name='Count')

# Pivot for heatmap
heatmap_data = affinity.pivot(index='Age', columns='Product_Category', values='Count').fillna(0)

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Product Affinity by Age Group')
plt.ylabel('Age Group')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
```
